```python
tickers = pd.read_csv('ASX Tickers.csv')
stocks = pd.DataFrame([['Ticker', 'Name', 'Sector', 'Industry', 'Price', 'Market Capitalisation', 'Shares Outstanding', '52 Week Change', 'PE Ratio (TTM)', 'Forward PE Ratio', 'PEG Ratio (5YR Expected)', 'Dividend Yield', 'PB Ratio', 'PS Ratio (TTM)', 'EV/Revenue', 'EV/EBITDA', 'Profit Margin']])
stocks=stocks.T.set_index(0).T
stocks['Ticker'] = tickers['Ticker']
stocks.head()
```

```python
#Exclusion List
exclude = ["SIX", "D13"]
```

```python
for i in range(1765,len(stocks['Ticker'])):
#for i in range(0,1):
    try:
        ticker = yf.Ticker(stocks['Ticker'].iloc[i] + ".AX")
        stocks['Name'].iloc[i] = ticker.info['longName']
    except KeyError:
        print('Ticker ' + stocks['Ticker'].iloc[i] +' not found: KeyError')
        exclude.append(stocks['Ticker'].iloc[i])
        continue
    except ValueError:
        print('Ticker ' + stocks['Ticker'].iloc[i] +' not found: ValueError')
        exclude.append(stocks['Ticker'].iloc[i])
        continue
    except IndexError:
        print('Ticker ' + stocks['Ticker'].iloc[i] +' not found: IndexError')
        exclude.append(stocks['Ticker'].iloc[i])
        continue
    ticker = yf.Ticker(stocks['Ticker'].iloc[i] + ".AX")
    stocks['Name'].iloc[i] = ticker.info['longName']
    stocks['Sector'].iloc[i] = ticker.info['sector']
    stocks['Industry'].iloc[i] = ticker.info['industry']
    stocks['Price'].iloc[i] = ticker.info['previousClose']
    stocks['Market Capitalisation'].iloc[i] = ticker.info['marketCap']
    stocks['Shares Outstanding'].iloc[i] = ticker.info['sharesOutstanding']
    stocks['52 Week Change'].iloc[i] = ticker.info['52WeekChange']
    if ticker.info['trailingEps'] == None or ticker.info['trailingEps'] == 0 or ticker.info['previousClose'] == None:
        stocks['PE Ratio (TTM)'].iloc[i] = "NaN"
    else:
        stocks['PE Ratio (TTM)'].iloc[i] = ticker.info['previousClose']/ticker.info['trailingEps']
    stocks['Forward PE Ratio'].iloc[i] = ticker.info['forwardPE']
    stocks['PEG Ratio (5YR Expected)'].iloc[i] = ticker.info['pegRatio']
    stocks['Dividend Yield'].iloc[i] = ticker.info['dividendYield']
    stocks['PB Ratio'].iloc[i] = ticker.info['priceToBook']
    stocks['PS Ratio (TTM)'].iloc[i] = ticker.info['priceToSalesTrailing12Months']
    stocks['EV/Revenue'].iloc[i] = ticker.info['enterpriseToRevenue']
    stocks['EV/EBITDA'].iloc[i] = ticker.info['enterpriseToEbitda']
    stocks['Profit Margin'].iloc[i] = ticker.info['profitMargins']
            
```
